```python
#import pandas
import pandas as pd

#import language data
languages_df = pd.read_csv("RAW/CSV_Languages_copy.csv")



#import population data

population_df = pd.read_csv("RAW/Chicago_Census2010populations_copy.csv")

#append populations 

frames = [languages_df, population_df]

working_df = pd.concat (frames, axis = 1)

#take out superfluous community areas

final_languages_df = working_df.drop(labels=["Community Areas", "Num"], axis=1)

#inspect first 20 lines
display(final_languages_df.head(20))
```

Step 3 (5 pts.): Use numpy and the cleaned data to perform several calculations relevant to your final project. Explain each calculation by including comments within your code. Your output should clearly state what each calculation is, i.e. "The mean height of those in the data is X"

```python
#import numpy

import numpy as np

#sum number isolated residents per community area

# drop the columns which are not language columns
sum_df = final_languages_df.drop(labels = ["Community Area", "Community Area Name", "PREDOMINANT NON-ENGLISH LANGUAGE (%)", "2010"], axis=1)

#sum sum_df
some_sums = np.sum(sum_df, axis = 1)


# append to final langauges 
frames = [some_sums, final_languages_df]

sum_final_df = pd.concat (frames, axis = 1)
# give the new column a name
sum_final_df.rename(columns={sum_final_df.columns[0]:'Subtotal'}, inplace=True)

#divide sum_final_df

divide_sum = np.divide(sum_final_df.loc[:, 'Subtotal'], pd.to_numeric(sum_final_df.loc[:, '2010']))

#percentage
percent_sum = np.multiply(divide_sum, 100)

# append to final langauges 
frames = [percent_sum, sum_final_df]

percent_final_df = pd.concat (frames, axis = 1)

# give the new column a name
old_columns = pd.Series(percent_final_df.columns)
old_columns[0] = 'Percent'
percent_final_df.columns = old_columns
#percent_final_df.rename(columns={percent_final_df.columns[0]:'Percent'}, inplace=True)

display(percent_final_df)
    
    
```

```python
#make an array of community areas
comm_areas = np.loadtxt('RAW/CSV_Languages_copy.csv', delimiter=',', skiprows = 1, usecols= (1), dtype = str)

#make an array of chinese households
chi_house = np.loadtxt('RAW/CSV_Languages_copy.csv', delimiter=',', skiprows = 1, usecols= (7))



#import matplotlib
import matplotlib.pyplot as plt

#initializing the figure
fig, ax = plt.subplots()

#y is the number of community areas
y_pos = np.arange(len(comm_areas))

#creating the bar chart
ax.barh(y_pos, chi_house, align='center', color='teal')

#labeling bar chart
ax.set_yticks(y_pos)
ax.set_yticklabels(comm_areas)
ax.invert_yaxis()
ax.set_xlabel('Number of Chinese residents')
ax.set_title('Number of linguistically-isolated residents by community area')
fig.set_figheight(18)

plt.show()

#export chart
fig.savefig('NumberChart.png', bbox_inches='tight', pad_inches=0.1)
```

```python

comm_areas = percent_final_df.loc[:, 'Community Area Name']
percent = percent_final_df.loc[:, 'Percent']


#initializing the figure
fig, ax = plt.subplots()

#y is the number of community areas
y_pos = np.arange(len(comm_areas))

#creating the bar chart
ax.barh(y_pos, percent, align='center', color='red')

#labeling bar chart
ax.set_yticks(y_pos)
ax.set_yticklabels(comm_areas)
ax.invert_yaxis()
ax.set_xlabel('Percentage of residents who don\'t speak English \'very well\'')
ax.set_title('Percentage of linguistically-isolated residents by community area')
fig.set_figheight(18)

plt.show()

#export chart
fig.savefig('PercentageChart.png', bbox_inches='tight', pad_inches=0.1)

```
